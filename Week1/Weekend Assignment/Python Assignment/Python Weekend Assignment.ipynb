```python
########## 3.

def isBalanced(brackets):
    sb_open = 0
    sb_close = 0
    cb_open = 0
    cb_close = 0
    bb_open = 0 
    bb_close = 0
    for i in range(0, len(brackets)):   
        if(brackets[i] == '('):
            sb_open += 1
        if (brackets[i] == ')'):
            sb_close += 1
        if (brackets[i] == '{'):
            cb_open +=1
        if (brackets[i] == '}'):
            cb_close +=1
        if (brackets[i] == '['):
            bb_open += 1
        if (brackets[i] == ']'):
            bb_close += 1
            
    if ((sb_open == sb_close) and (cb_open == cb_close) and (bb_open == bb_close)):
        return 'YES'
    else:
        return 'No'

s = input('[+] Enter bracket combinations: ')
condition = isBalanced(s)

print(f'[+] YES' if condition=='YES' else '[-] NO')
```

```python
############# 4.

import math
class Complex:
    def __init__(self,real=0,imaginary=0):
        self.re = real
        self.im = imaginary
            
    def __str__(self):
        return(f'{self.re} {"+" if self.im >=0 else "-"} {abs(self.im)}i')
    
    
    ##### ARITHMETIC OPERATIONS 
    
    def __add__(self,other):
        return (f'\n[+] Addition: {Complex(self.re + other.re, self.im + other.im )}')

    def __sub__(self,other):
        return (f'\n[+] Subtraction: {Complex(self.re - other.re, self.im - other.im )}')
    
    
    def __mul__(self,other):
        re = self.re * other.re - self.im * other.im
        im = self.im * other.re + self.re * other.im
        return (f'\n[+] Multiplication: {Complex(re,im)}')
    
    @property
    def conjugate(self):
        return (f'\n[+] Conjugate of ({self}): {Complex(self.re,-self.im)}')
    
    def __truediv__(self,other):      
        r = (other.re**2 + other.im**2)
        return (f'\n[+] Division: {Complex((self.re*other.re-self.im*other.im)/r, (self.im*other.re+self.re*other.im)/r)}')
    
    def __abs__(self):
        return (f'    Absolute value of ({self}): {math.sqrt(self.re**2+self.im**2)}')
        

    #### RELATIONAL OPERATIONS
    
    def __eq__(self,other):
        return self.re == other.re and self.im == other.im
    
    def __lt__(self,other):
        ab1 = math.sqrt(self.re**2+self.im**2)
        ab2 = math.sqrt(other.re**2+other.im**2)
        return (ab1 < ab2)

    def __gt__(self,other):
        ab1 = math.sqrt(self.re**2+self.im**2)
        ab2 = math.sqrt(other.re**2+other.im**2)
        return (ab1 > ab2)
    

c1 = Complex(4,15)
c2 = Complex(2, 6)

print(c1)
print(c2)

print(c1+c2)

# c1+=c2

print(c1-c2)

print(c1*c2)

print(c1.conjugate)

print(c1/c2)



print(f'\n\n[+] Equality of ({c1}) and ({c2}):',c1==c2)
print(f'[+] Equality of({c1}) and ({c1}):',c1==c1)

print(f'\n[+] ({c1}) is Less than ({c2}):',c1<c2)
print(abs(c1))
print(abs(c2))

print(f'\n[+] ({c1}) is Greater than ({c2}):',c1>c2)
```

```python
########### 5.
''' a. List(“1234”) works same as list(“1234”)   
    b. Adding 2 lists, multiple ways of instantiating using @classmethod.
'''

class List:
    def __init__(self, word):
        self.lst = "["
        for i in range(len(word)-1):
            self.lst = self.lst + "'" + word[i] + "'" + ", "
        self.lst = self.lst + "'" + word[-1] + "'"  + ']'
    
    def __str__(self):
        return self.lst
    
    def __add__(self, other):
        try:
            return (self.lst[:-1] + ', ' + other.lst[1:])
        except:
            return (self.lst[:-1] + ', ' + other[1:])
        
    
    @classmethod
    def makeList(self, word):
        if str(word)[0] != '[':
            raise ValueError("[-] Invalid !")
        if str(word)[-1] != ']':
            raise ValueError("[-] Invalid !")
        self.lst = str(word)
        return self.lst
    


a = List("1234")
print('[+] Required list:',a)

b = list("1234") + [1,2,3,4]
print('[+] Addition of lists:',b)

```

```python
########### 6. 
########### a. with iterator class, __iter__ and __next__

class Range:
    def __init__(self, *args):
        if len(args) == 3:
            self.start = args[0]
            self.end = args[1]
            self.diff = args[2]
        elif len(args) == 2:
            self.start = args[0]
            self.end = args[1]
            self.diff = 1
        elif len(args) == 1:
            self.start = 0
            self.end = args[0]
            self.diff = 1
        else:
            raise ValueError("[-] Invalid Format !")
    
    def __iter__(self):
        self.num = self.start - self.diff
        return self

    def __next__(self):
        self.num += self.diff
        if self.num < self.end:
            return self.num
        else:
            raise StopIteration
for i in Range(1,20,1):
    print(i, end = " ")
```

```python
######### 6. 
######### b.Generator

def Range(*args):
    if len(args) == 3:
        start = args[0]
        end = args[1]
        diff = args[2]
    elif len(args) == 2:
        start = args[0]
        end = args[1]
        diff = 1
    elif len(args) == 1:
        start = 0
        end = args[0]
        diff = 1
    else:
        raise ValueError("[-] Invalid Format !")
    result = start
    while result < end:
        yield result
        result += diff

for i in Range(0,11,2):
    print(i, end = " ")
```

```python
########## 6. 
########## c.Generator comprehension"""

def while_generator(start, end, diff):
    while start < end:
        yield start
        start += diff
        
def Range(*args):
    if len(args) == 3:
        start = args[0]
        end = args[1]
        diff = args[2]
    elif len(args) == 2:
        start = args[0]
        end = args[1]
        diff = 1
    elif len(args) == 1:
        start = 0
        end = args[0]
        diff = 1
    else:
        raise ValueError("[-] Invalid Format !")
    lst = [i for i in while_generator(start, end, diff)]
    return lst

for i in Range(1,11,2):
    print(i, end = " ")
```

```python
############# 2.
# Create a table named 'users'
# -- columns -> id, name, dob, profession 


import psycopg2

try:
    connection = psycopg2.connect(
        user="postgres",
        password="admin",
        host="localhost",
        port="5432",
        database="postgres"
    )
    
    cursor = connection.cursor()
    
    query = '''
    CREATE TABLE IF NOT EXISTS users(
        id INT generated always as identity primary key,
        name VARCHAR(30) NOT NULL,
        dob date,
        profession VARCHAR(100) NOT NULL
    )'''
    cursor.execute(query)
    connection.commit()
      
    
except Exception as e:
    print("[-] Exception Occurred:", e)

finally:
    cursor.execute("SELECT version();")
    print(cursor.fetchone())
    print("[+] Executed !")
    cursor.close()
    connection.close()
    print("[+] Connection Closed !")


```

```python
############ 3. 
# Create table names `address`
#    -- Columns -> id, user_id (FK -> users), permanent_address, temporary_address


import psycopg2

try:
    connection = psycopg2.connect(
        user="postgres",
        password="admin",
        host="localhost",
        port="5432",
        database="postgres"
    )
    
    cursor = connection.cursor()
    
    query = '''
    CREATE TABLE IF NOT EXISTS address(
        id  INT generated always as identity primary key,
        user_id int,
        permanent_address VARCHAR(50) NOT NULL,
        temporary_address VARCHAR(50),
        constraint fk_user_id foreign key (user_id) references users(id) 
    )'''
    cursor.execute(query)
    connection.commit()
      
    
except Exception as e:
    print("[-] Exception Occurred:", e)

finally:
    cursor.execute("SELECT version();")
    print(cursor.fetchone())
    print("[+] Executed !")
    cursor.close()
    connection.close()
    print("[+] Connection Closed !")
```
