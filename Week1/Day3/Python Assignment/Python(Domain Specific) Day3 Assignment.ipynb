```python
# 1. 

import psycopg2

try:
    connection = psycopg2.connect(
        user="postgres",
        password="admin",
        host="localhost",
        port="5432",
        database="postgres"
    )
    
    cursor = connection.cursor()
    
    query1 = '''
    CREATE TABLE IF NOT EXISTS DoctorSpecialization(
        id serial PRIMARY KEY,
        specialization_type varchar(50) NOT NULL
    )'''
    
    query2 = '''
    CREATE TABLE IF NOT EXISTS Doctor(
        id serial PRIMARY KEY,
        name VARCHAR(50) NOT NULL,
        specialization INT NOT NULL,
        phone_number VARCHAR(15),
        FOREIGN KEY(specialization) REFERENCES 
        DoctorSpecialization(id) 
        )'''
    
    query3 = '''
    CREATE TABLE IF NOT EXISTS Patient(
        id SERIAL PRIMARY KEY,
        name VARCHAR(50) NOT NULL,
        date_of_birth VARCHAR(30) NOT NULL,
        gender VARCHAR(10) NOT NULL
    )'''
    
   
    query4 = '''
    CREATE TABLE IF NOT EXISTS Appointment(
        id serial PRIMARY KEY,
        doctor_id INT NOT NULL,
        patient_id INT NOT NULL,
        fee INT NOT NULL,
        diagnosis VARCHAR(50),
        FOREIGN KEY(doctor_id) REFERENCES
        Doctor(id),
        FOREIGN KEY(patient_id) REFERENCES
        Patient(id)
    )'''

    cursor.execute(query1)
    cursor.execute(query2)
    cursor.execute(query3)
    cursor.execute(query4)
    
    connection.commit()
    
except Exception as e:
    print("[-] Exception Occurred:", e)

finally:
    cursor.execute("SELECT version();")
    print(cursor.fetchone())
    print("[+] Executed !")
    cursor.close()
    connection.close()
    print("[+] Connection Closed !")

```

```python
# 7.

import psycopg2

try:
    connection = psycopg2.connect(
        user="postgres",
        password="admin",
        host="localhost",
        port="5432",
        database="postgres"
    )
    
    cursor = connection.cursor()
    
    select = ''' SELECT * FROM Doctor '''
    cursor.execute(select)
    print('[+] Doctor Details: ',cursor.fetchall(),'\n')
    
    del_phychiatrist = '''
    DELETE FROM Doctor
    WHERE id IN (
        SELECT id from DoctorSpecialization
        WHERE specialization_type = 'Psychiatrist')
    '''
    cursor.execute(del_phychiatrist)
    
    
    select = ''' SELECT * FROM Doctor '''
    cursor.execute(select)
    print('[+] Doctor Details after removing Psychiatrist: ',cursor.fetchall(),'\n')
    
    connection.commit()
    
except Exception as e:
    print("[-] Exception Occurred:", e)

finally:
    cursor.execute("SELECT version();")
    print(cursor.fetchone())
    print("[+] Executed !")
    cursor.close()
    connection.close()
    print("[+] Connection Closed !")

```
